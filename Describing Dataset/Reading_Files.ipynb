# Introduction to Python: Reading Files

```
import pandas as pd
import numpy as ny
from numpy import percentile
california = pd.read_csv ('/content/sample_data/california_housing_train.csv')

california.describe()

```

```
mean=ny.mean (california)

std=ny.std(california)

qhma1=percentile(california['housing_median_age'],[25])
qtr1=percentile(california['total_rooms'],[25])
qtb1=percentile(california['total_bedrooms'],[25])
qp1=percentile(california['population'],[25])
qh1=percentile(california['households'],[25])
qmi1=percentile(california['median_income'],[25])
qmhv1=percentile(california['median_house_value'],[25])

qhma2=percentile(california['housing_median_age'],[50])
qtr2=percentile(california['total_rooms'],[50])
qtb2=percentile(california['total_bedrooms'],[50])
qp2=percentile(california['population'],[50])
qh2=percentile(california['households'],[50])
qmi2=percentile(california['median_income'],[50])
qmhv2=percentile(california['median_house_value'],[50])

qhma3=percentile(california['housing_median_age'],[75])
qtr3=percentile(california['total_rooms'],[75])
qtb3=percentile(california['total_bedrooms'],[75])
qp3=percentile(california['population'],[75])
qh3=percentile(california['households'],[75])
qmi3=percentile(california['median_income'],[75])
qmhv3=percentile(california['median_house_value'],[75])

print("NILAI MEAN")
print(mean)

print("")

print("NILAI STANDARD DEVIASI")
print(std)

print("")

print("NILAI MIN HOUSING MEDIAN AGE = ", california['housing_median_age'].min())
print("nilai MIN TOTAL ROOMS = ", california['total_rooms'].min())
print("nilai MIN TOTAL BEDROOMS = ", california['total_bedrooms'].min())
print("nilai MIN POPULATION = ", california['population'].min())
print("nilai MIN HOUSEHOLDS = ", california['households'].min())
print("nilai MIN MEDIAN INCOME = ", california['median_income'].min())
print("nilai MIN MEDIAN HOUSE VALUE = ", california['median_house_value'].min())

print("")

print("NILAI MAX HOUSING MEDIAN AGE = ", california['housing_median_age'].max())
print("nilai MAX TOTAL ROOMS = ", california['total_rooms'].max())
print("nilai MAX TOTAL BEDROOMS = ", california['total_bedrooms'].max())
print("nilai MAX POPULATION = ", california['population'].max())
print("nilai MAX HOUSEHOLDS = ", california['households'].max())
print("nilai MAX MEDIAN INCOME = ", california['median_income'].max())
print("nilai MAX MEDIAN HOUSE VALUE = ", california['median_house_value'].max())

print("")

print("NILAI Q1 HOUSING MEDIAN AGE =",qhma1, "NILAI Q2 HOUSING MEDIAN AGE =",qhma2, "NILAI Q3 HOUSING MEDIAN AGE =",qhma3)
print("NILAI Q1 TOTAL ROOM =",qtr1,"NILAI Q2 TOTAL ROOM =",qtr2, "NILAI Q3 TOTAL ROOM =",qtr3)
print("NILAI Q1 TOTAL BEDROOM =",qtb1, "NILAI Q2 TOTAL BEDROOM =",qtb2, "NILAI Q3 TOTAL BEDROOM =",qtb3)
print("NILAI Q1 POPULATION =",qp1, "NILAI Q2 POPULATION =",qp2, "NILAI Q3 POPULATION =",qp3)
print("NILAI Q1 HOUSEHOLDS =",qh1 , "NILAI Q2 HOUSEHOLDS =",qh2, "NILAI Q3 HOUSEHOLDS =",qh3)
print("NILAI Q1 MEDIAN INCOME =",qmi1, "NILAI Q2 MEDIAN INCOME =",qmi2, "NILAI Q3 MEDIAN INCOME =",qmi3 )
print("NILAI Q1 MEDIAN HOUSE VALUE =",qmhv1, "NILAI Q2 MEDIAN HOUSE VALUE =",qmhv2, "NILAI Q3 MEDIAN HOUSE VALUE =",qmhv3 )
```
